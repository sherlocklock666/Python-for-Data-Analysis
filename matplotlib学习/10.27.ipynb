```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import sympy
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.sin(x)*np.exp(-x**2/20)
fig, axes = plt.subplots(1,2,figsize=(12,3))
axes[0].plot(x, y , lw=2)
axes[0].set_title("主次刻度")
axes[0].xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))#主刻度包括4个单位,显示数字
axes[0].xaxis.set_minor_locator(mpl.ticker.MultipleLocator(1.5))
axes[0].yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
axes[0].yaxis.set_minor_locator(mpl.ticker.FixedLocator([-0.2,0,0.2]))

# axes[1].plot(x, y , lw=2)
# axes[1].set_title("多倍刻度")
# axes[1].xaxis.set_major_locator(mpl.ticker.MultipleLocator(4))#主刻度包括4个单位,显示数字
# axes[1].xaxis.set_minor_locator(mpl.ticker.MultipleLocator(1))
# axes[1].yaxis.set_major_locator(mpl.ticker.MultipleLocator(0.5))
# axes[1].yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.25))
```
